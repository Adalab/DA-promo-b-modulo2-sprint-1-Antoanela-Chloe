Sacad los valores √∫nicos y su frecuencia de las variables num√©ricas. Almacenad los resultados en un diccionario. Uno para cada variable.

```python
dic_values = {}
for column in df.columns:
    dic_values.setdefault(column, df[column].unique())

dic_values
```

```python
dic_counts = {}
for column in df.columns:
    dic_counts.setdefault(column, df[column].value_counts())

dic_counts
```

Cambiad el nombre de las columnas para que todas est√©n en min√∫sculas y reemplazad los espacios por _.

```python
columnas_nuevas = {col :  col.lower().replace(' ', '_') for col in df.columns}
```

```python
df.rename(columns = columnas_nuevas, inplace = True)
df.head(2)
```

Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, Investigator or Source, pdf, href formula.

```python
df_drop = df.drop(["unnamed:_23", "case_number.1", "unnamed:_22", "case_number.2", "original_order", "investigator_or_source", "pdf", "href_formula"], axis = 1)
df_drop.head(3)
```

Establece la columna Case Number como √≠ndice del DataFrame.

```python
df_drop.set_index(["case_number"], inplace= True)
df_drop.head(3)
```
