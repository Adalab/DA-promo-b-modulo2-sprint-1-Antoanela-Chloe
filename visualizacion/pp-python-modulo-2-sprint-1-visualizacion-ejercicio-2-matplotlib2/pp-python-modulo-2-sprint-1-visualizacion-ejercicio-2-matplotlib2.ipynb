
1. Ayer creamos un par de gráficas para USA y España, donde visualizamos el número de ataques en cada país a lo largo de los años y el tipo de los ataques. Hoy tendréis que hacer dos gráficas con subplots:

En la primera debéis mostrar los tipos ataques en USA y España, cada subplot corresponderá a un país. Debéis:
* Poner nombre a los ejes de cada uno de los subplots
* Poner un títulos a cada uno de los subplots
* Poner el color de las líneas en negro.
* Establecer el tamaño de la gráfica a 15 x 5
* Ponerle marcadores a las gráficas
* No incluir leyenda


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('attacks_modificado.csv', index_col= 0)
```

```python
# Volvemos a filtar nuestro df por paises del año 2000 en adelante
usa2000 = df[(df['country'] == 'USA') & (df['year']>= 2000)]
spain2000 = df[(df['country'] == 'SPAIN') & (df['year']>=2000)]
```

```python
## ACLARACIÓN: Al igual que el el pair anterior, creemos que la gráfica que mejor ilustra lo que se pide es la gráfica de barras. 
# Por eso el apartado que pide que añadamos marcadores a las gráficas lo haremos en el ejercicio 2, donde hemos aplicado una gráfica lineal.
```

```python
x0 = spain2000['type'].value_counts().index
y0 = spain2000['type'].value_counts().values

x1 = usa2000['type'].value_counts().index
y1 = usa2000['type'].value_counts().values
```

```python
fig, ax = plt.subplots(1,2, figsize= (15, 5))
ax[0].set_xlabel(xlabel = 'Tipo de ataques')
ax[0].set_ylabel(ylabel = 'Cantidad de ataques')
ax[0].set_title('Tipos de ataques de tiburón en España desde el año 2000')
ax[0].bar(x0, y0, color = 'green', edgecolor = 'black')

ax[1].set_xlabel(xlabel = 'Tipo de ataques')
ax[1].set_ylabel(ylabel = 'Cantidad de ataques')
ax[1].set_title('Tipos de ataques de tiburón en EEUU desde el año 2000')
ax[1].bar(x1, y1, color = 'green', edgecolor = 'black')
```


En la segunda debéis mostrar los ataques por año para cada uno de los países.
* De nuevo tendréis que ponerle nombre a los ejes y título a cada gráfica.
* En este caso el subplot deberá tener dos filas y una columna
* No incluir leyenda

```python
x0 = spain2000['year'].value_counts().index.sort_values()
y0 = spain2000['year'].value_counts().values

x1 = usa2000['year'].value_counts().index.sort_values()
y1 = usa2000['year'].value_counts().values
```

```python
fig, ax = plt.subplots(2,1, figsize= (10, 15))
ax[0].set_xlabel(xlabel = 'Año')
ax[0].set_ylabel(ylabel = 'Cantidad de ataques')
ax[0].set_title('Numero de ataques de tiburón por año en España')
ax[0].plot(x0, y0, color = 'black', marker = 'o')

ax[1].set_xlabel(xlabel = 'Año')
ax[1].set_ylabel(ylabel = 'Cantidad de ataques')
ax[1].set_title('Numero de ataques de tiburón por año en EEUU')
ax[1].plot(x1, y1, color = 'black', marker = 'o')
```

2. Cread un único gráfico donde juntéis las gráficas de los tipos de ataques en una sola. Debéis:
Poner leyenda a la gráfica
Cada línea debe ir en un color
Ponerle nombre a los ejes y a la gráfica.
Poner marcadores a las lineas.

```python
tipos_0 = spain2000['type'].value_counts().index
tipos_1 = usa2000['type'].value_counts().index

x0 = np.arange(len(tipos_0), dtype = float)
x1 = np.arange(len(tipos_1), dtype = float)

y0 = spain2000['type'].value_counts().values
y1 = usa2000['type'].value_counts().values
```

```python
labels = list(usa2000['type'].value_counts().index)
labels
```

```python
fig, ax = plt.subplots(figsize= (15, 5))

spain = ax.bar(x0 - (0.35/2), y0, width = 0.35, color = 'green', edgecolor = 'black', label = 'España')
usa = ax.bar(x1 + (0.35/2), y1, width = 0.35, color = 'orange', edgecolor = 'black',  label ='EEUU' )

ax.set_xticks(list(x1)); ax.set_xticklabels(list(tipos_1))

ax.legend()

ax.set_xlabel(xlabel = 'Tipo de ataques')
ax.set_ylabel(ylabel = 'Cantidad de ataques')

fig.tight_layout()

plt.title('Tipos de ataques de tiburón en España y EE.UUdesde el año 2000')
plt.show()
```
